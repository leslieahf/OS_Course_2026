```python
import os
import stat

def list_directory_contents(path):

    # To verify if the path exists and if it is a directory
    if not os.path.exists(path):
        print(f"Error : The path '{path}' doesn't exist")
        return
    if not os.path.isdir(path):
        print(f"Error : '{path}' is not a directory")
        return

    print("\n")

    # To list all files inside the directory
    file_list = os.listdir(path)
    print(f"All the files inside '{path}' are :")
    print(file_list)

    print("\n")

    # To retrieve and print characteristics of each file
    print("Its characteristics are : \n")
    print(f"{'File name':<30} | {'Size (bytes)':<15} | {'Permissions'}")
    print("-" * 65)

    try:
        # To iterate through the directory
        for file in os.scandir(path):
                # To retrieve statistics of the file
                info = file.stat()
                # File size (bytes)
                size = info.st_size
                # Conversion of file permissions (-rw-r--r--)
                permissions = stat.filemode(info.st_mode)

                print(f"{file.name:<30} | {size:<15} | {permissions}")

    except PermissionError:
        print("Error : Permission denied to access this file")

if __name__ == "__main__":
    # Ask the user for directory path
    dir_path = input("Enter the directory path to analyze : ")
    list_directory_contents(dir_path)
```
