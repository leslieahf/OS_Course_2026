```python
# Script that print basic system information

%%bash

# To print operating system name (s for system name)
echo "OS: $(uname -s)"
# To print kernel version (r for kernel release)
echo "Kernel Version: $(uname -r)"
# To print the current logged-in user
# In Google Colab, this user is 'root'
echo "Current User: $(whoami)"
# To print the current working directory
echo "Working Dir:  $(pwd)"
```
